```python
import pandas as pd
import numpy as np
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import altair as alt

alt.renderers.enable('notebook')
alt.renderers.enable('html')
spacy_nlp = spacy.load('en_core_web_sm')
```

```python
df = pd.read_csv("test_reviews_small.csv")
df
```

```python
def basic_tokenizer(sentence):
    doc = spacy_nlp(sentence)
    tokens = [token.text for token in doc]
    return tokens
```

```python
def clean_text(text):
    doc = spacy_nlp(text)
    toks = [token.text for token in doc if not (token.is_stop or token.is_punct)]
    return " ".join(toks)
```

```python
df['review'] = df['review'].str.lower()
df['review'] = df['review'].map(clean_text)
df['review']
```

```python
vectorizer = TfidfVectorizer(tokenizer=basic_tokenizer)
tfidf_vectors = vectorizer.fit_transform(df['review'])
#df['review-tfidf'] = [v.toarray() for v in tfidf_vectors]
#df['review-tfidf'] = [list(v) for v in tfidf_vectors.A]
df['review-tfidf'] = list(tfidf_vectors`)
df['review-tfidf']
```

```python
print(df['review-tfidf'].shape)
tfidf_2d = np.stack(df['review-tfidf'], axis=0)
print(tfidf_2d.shape)
print(tfidf_2d)
```

```python
pca = PCA(n_components=10)
pca_vectors = pca.fit_transform(tfidf_2d)
df['review-pca'] = pca_vectors.tolist()
# shape should be (73,10) because taking only first 10 pca components
print('shape of pca vectors is: ', pca_vectors.shape)
print('first row of pca vectors is: ', df['review-pca'][0])
```

```python
df['pca_0'] = df['review-pca'].map(lambda x: x[0])
df['pca_0']
```

```python
df['pca_1'] = df['review-pca'].map(lambda x: x[1])
df['pca_1']
```

```python
df.iloc[0]
```

```python
pca_data = []
c = 0
for index, row in df[['pca_0', 'pca_1']].iterrows():
    if c < 10:
        print(index, row['pca_0'], row['pca_1'])
    c += 1
    pca_data.append([row['pca_0'], row['pca_1']])
pca_data[:5]
```

```python
kmeans = KMeans(n_clusters=6, n_init=10, verbose=0).fit(tfidf_2d)
cluster_preds = kmeans.predict(tfidf_2d)
print(cluster_preds)
print(cluster_preds.shape)
```

```python
df['review-clusters'] = [str(v) for v in cluster_preds]
df.iloc[0]
```

```python
print(kmeans.cluster_centers_.shape)
kmeans.cluster_centers_
```

```python
alt.Chart(df).mark_circle().encode(
    alt.X('pca_0'),
    alt.Y('pca_1'),
    color='review-clusters'
)
```

```python
# apply sentiment operator to review column
sid = SentimentIntensityAnalyzer()
df['review-sentiment'] = [sid.polarity_scores(r)['compound'] for r in df['review']]
df['review-sentiment']
```

```python
alt.Chart(df).mark_circle().encode(
    alt.X('pca_0'),
    alt.Y('pca_1'),
    color=alt.Color(field='review-sentiment', type='quantitative', scale=alt.Scale(range=["crimson", "blue"])),
)
```
